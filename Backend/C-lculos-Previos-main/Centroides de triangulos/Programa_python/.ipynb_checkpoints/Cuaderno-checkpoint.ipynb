```python
import numpy as np
import trimesh
import os
import calculo_centroides #contiene las coordenadas de los puntos críticos
import math 


print(filename)
mesh_SES =trimesh.load('prueba.stl')
#Paracentros_esferas=calculo_centroides.getCenter()


#compute_rhino3d.Util.authToken = AUTH_TOKEN_FROM (rhino3d.com/compute/login)

centros_esferas=calculo_centroides.getCenter()

def computeCR(radii,delta,centers):
#radii:radio de la esfera de los CR
#delta: cantidad a generar
#centers: centros de puntos críticos
#n: division de cada rayo
    phi=np.linspace(0, 2*math.pi, delta) #tamano 2pi*10^2=>1256->36^2 =>1296, 35^2=>1225
    theta=np.linspace(0, 2*math.pi, delta)
    ray_directions=list()
    ray_origins=list()
    counter=0 
    cr=list()
    for item in centers:
        ray_origins_item=list()
        ray_directions_item=list()
        for i in phi:
            for j in theta:
                #Se calcula el término del rayo y se le suma el centro ya que no está en el origen 0,0,0
                x=radii*math.sin(i)*math.cos(j)+item[0]
                y=radii*math.sin(i)*math.sin(j)+item[1]
                z=radii*math.cos(i)+item[2]
                ray_directions_item.append([x,y,z])
                ray_origins_item.append(item)
                #Para calcular donde está el SES
                #print("hits",hits)
        ray_directions.append(ray_directions_item)
        ray_origins.append(ray_origins_item)
        cr.append([counter,item,ray_directions_item])
        counter=counter+1
    print("fin calculo CR")    
    return ray_origins,ray_directions,cr

origins,ends,cr=computeCR(3,10,centros_esferas)
n=8
cont=0
cr_segments=list()
for item in cr:    
    #item[0]=indice, item[1]=coord punto crítico, item[2] coordenada fin de cr
    origen=item[1]
    finales=item[2]
    for rayo in finales:
        # print("count",cont)
        vector=np.linspace(origen,rayo,n)
        hits=mesh_SES.contains(vector)
        cr_segments.append(hits)
        # print("hits",hits)

        # hits=mesh_SES.ray.intersects_any(ray_origins=vector[0:n-1],ray_directions=vector[1:n])
        # cr_segments.append([item[0],hits])
        # locations, index_ray, index_tri = mesh_SES.ray.intersects_location(ray_origins=vector[0:n-1],
        # ray_directions=vector[1:n])
 
```
